#### 1. Import pandas library

```python
import pandas as pd
```

#### 2. Load data (users.csv, posts.csv)

```python
data_users = pd.read_csv('users.csv')
data_users.head(2)
```

```python
data_post = pd.read_csv('posts.csv')
data_post.head(2)
```

#### 3. Rename id column to user_id

```python
# rename id column of first data set: data_users
data_users = data_users.rename(columns={'id': 'user_id'})
data_users.head(2)
```

#### 4. Rename id column to post_id and owner_user_id to user_id

```python
data_post = data_post.rename(columns={'id':'post_id', 'owner_user_id':'user_id'})
data_post.head(2)
```

#### 5. Define new dataframes for users and posts with the following selected columns:
    **users columns**: user_id, reputation,views,up_votes,down_votes
    **posts columns**: post_id, score,user_id,view_count,comment_count

```python
desired_cols = ['user_id','reputation', 'views', 'up_votes', 'down_votes']
subset_users = data_users[desired_cols]
subset_users.head(2)
```

```python
desired_cols_post = ['post_id', 'score', 'user_id', 'view_count', 'comment_count']
subset_post = data_post[desired_cols_post]
subset_post.head(2)
```

#### 6. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
data_merge =  pd.merge(data_users, data_post)
```

```python
print(data_users.shape)
print(data_post.shape)
print(data_merge.shape)
```

```python
data_merge
```

#### 7. How many missing values do you have in your merged dataframe? 

```python
data_merge.isnull().sum()
```

```python
pd.isnull(data_merge).values.ravel().sum()
```

```python
# Clean nan values 
```

```python
data_merge.dropna(how='all')
```

```python
data_merge['about_me'] = data_merge['about_me'].fillna(0)
data_merge['profile_image_url'] = data_merge['profile_image_url'].fillna(0)
data_merge['website_url'] = data_merge['website_url'].fillna('Unknow')
data_merge['favorite_count'] =  data_merge['favorite_count'].fillna(0)
data_merge['age'] = data_merge['age'].fillna(data_merge['age'].mean())
```

There are no missing values
